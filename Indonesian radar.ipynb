```python
import pyart
from matplotlib import pyplot as plt
import numpy as np
%matplotlib inline
```

```python
filename = '/bm/gdata/srauniya/JEP100103001216.RAWXN3S'
radar = pyart.io.read(filename)
```

```python
print radar.fields.keys()
```

```python
print radar.metadata
```

```python
display = pyart.graph.RadarMapDisplay(radar)
fig = plt.figure(figsize=[12,5])
display.plot_ppi_map('reflectivity', resolution = 'l', vmax = 64, vmin  = 0)
m = display.basemap
m.drawrivers()
m.drawparallels(np.linspace(-8, -4, 5),labels=[1,0,0,0])
m.drawmeridians(np.linspace(102, 108, 5),labels=[0,0,0,1])
```
